```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from geopy.distance import geodesic as GD
import math
```

```python
df = pd.DataFrame()
num = 500
df['Number'] = range(num)
df['Gender'] = np.random.choice(['Male','Female','Nonbinary'], size=num, p=[0.47,0.47,0.06])
df['Chatbot'] = np.random.choice(['Yes', 'No'], size=num, p=[0.05, 0.95])
df['Height'] = np.random.randint(140, 200, num)
df['Age'] = np.random.randint(18, 35, num)
df['Education'] = np.random.choice(['High School','Bachelor','Master','PhD'], size=num, p=[0.4, 0.48, 0.1, 0.02])
df['E or I'] = np.random.choice(['E', 'I'], size=num, p=[0.5, 0.5])
df['S or N'] = np.random.choice(['S', 'N'], size=num, p=[0.5, 0.5])
df['T or F'] = np.random.choice(['T', 'F'], size=num, p=[0.5, 0.5])
df['J or P'] = np.random.choice(['J', 'P'], size=num, p=[0.5, 0.5])
df['Latitude'] = np.random.randint(-90, 90, num)
df['Longitude'] = np.random.randint(-180, 180, num)
for i in range(num):
    if df.loc[i,'Gender'] == 'Male':
        df.loc[i,'Gender Prefer'] = np.random.choice(['Male', 'Female', 'Nonbinary', 'All'], p=[0.1, 0.78, 0.02, 0.1])
    if df.loc[i,'Gender'] == 'Female':
        df.loc[i,'Gender Prefer'] = np.random.choice(['Male', 'Female', 'Nonbinary', 'All'], p=[0.78, 0.1, 0.02, 0.1])
    if df.loc[i,'Gender'] == 'Nonbinary':
        df.loc[i,'Gender Prefer'] = np.random.choice(['Male', 'Female', 'Nonbinary', 'All'], p=[0.1, 0.1, 0.7, 0.1])
df['Height Prefer'] = df['Height'] + np.random.randint(-10, 10, num)
df['Age Prefer'] = df['Age'] + np.random.randint(-2, 2, num)
df['Education Prefer'] = np.random.choice(['High School','Bachelor','Master','PhD'], size=num, p=[0.4, 0.48, 0.1, 0.02])
df['E or I Prefer'] = np.random.choice(['E', 'I'], size=num, p=[0.5, 0.5])
df['S or N Prefer'] = np.random.choice(['S', 'N'], size=num, p=[0.5, 0.5])
df['T or F Prefer'] = np.random.choice(['T', 'F'], size=num, p=[0.5, 0.5])
df['J or P Prefer'] = np.random.choice(['J', 'P'], size=num, p=[0.5, 0.5])
df['Distance Prefer'] = np.random.randint(0, 300, num)
df
```

```python
db = df.copy()
for i in range(len(db)):
    if db.loc[i,'Education'] == 'High School':
        db.loc[i,'Education'] = 0
    if db.loc[i,'Education'] == 'Bachelor':
        db.loc[i,'Education'] = 1
    if db.loc[i,'Education'] == 'Master':
        db.loc[i,'Education'] = 2
    if db.loc[i,'Education'] == 'PhD':
        db.loc[i,'Education'] = 3
    if db.loc[i,'E or I'] == 'E':
        db.loc[i,'E or I'] = 0
    else:
        db.loc[i,'E or I'] = 1
    if db.loc[i,'S or N'] == 'S':
        db.loc[i,'S or N'] = 0
    else:
        db.loc[i,'S or N'] = 1
    if db.loc[i,'T or F'] == 'T':
        db.loc[i,'T or F'] = 0
    else:
        db.loc[i,'T or F'] = 1
    if db.loc[i,'J or P'] == 'J':
        db.loc[i,'J or P'] = 0
    else:
        db.loc[i,'J or P'] = 1
    if db.loc[i,'Education Prefer'] == 'High School':
        db.loc[i,'Education Prefer'] = 0
    if db.loc[i,'Education Prefer'] == 'Bachelor':
        db.loc[i,'Education Prefer'] = 1
    if db.loc[i,'Education Prefer'] == 'Master':
        db.loc[i,'Education Prefer'] = 2
    if db.loc[i,'Education Prefer'] == 'PhD':
        db.loc[i,'Education Prefer'] = 3
    if db.loc[i,'E or I Prefer'] == 'E':
        db.loc[i,'E or I Prefer'] = 0
    else:
        db.loc[i,'E or I Prefer'] = 1
    if db.loc[i,'S or N Prefer'] == 'S':
        db.loc[i,'S or N Prefer'] = 0
    else:
        db.loc[i,'S or N Prefer'] = 1
    if db.loc[i,'T or F Prefer'] == 'T':
        db.loc[i,'T or F Prefer'] = 0
    else:
        db.loc[i,'T or F Prefer'] = 1
    if db.loc[i,'J or P Prefer'] == 'J':
        db.loc[i,'J or P Prefer'] = 0
    else:
        db.loc[i,'J or P Prefer'] = 1
db
```

```python
Like = []
Dislike = []
Rest = []
temp = []
for i in range(num):
    Like.append(temp)
    Dislike.append(temp)
for i in range(num):
    temp = []
    if db.loc[i,'Gender Prefer'] != 'All':
        for j in range(num):
            if j!=i and db.loc[i,'Gender Prefer']==db.loc[j,'Gender']:
                temp.append(j)
    else:
        for j in range(num):
            if j!=i:
                temp.append(j)
    Rest.append(temp)
```

```python
def factor(i,j,db):
    height = abs(db.loc[i,'Height Prefer']-db.loc[j,'Height'])
    age = abs(db.loc[i,'Age Prefer']-db.loc[j,'Age'])
    education = abs(db.loc[i,'Education Prefer']-db.loc[j,'Education'])
    ei = abs(db.loc[i,'E or I Prefer']-db.loc[j,'E or I'])
    sn = abs(db.loc[i,'S or N Prefer']-db.loc[j,'S or N'])
    tf = abs(db.loc[i,'T or F Prefer']-db.loc[j,'T or F'])
    jp = abs(db.loc[i,'J or P Prefer']-db.loc[j,'J or P'])
    distance = abs(db.loc[i,'Distance Prefer']-GD(np.array([db.loc[i,'Latitude'],db.loc[i,'Longitude']]),np.array([db.loc[j,'Latitude'],db.loc[j,'Longitude']])).km)
    return np.array([height,age,education,ei,sn,tf,jp,distance])
```

```python
def match(n, Like, Dislike, Rest, db):
    if db.loc[n,'Chatbot'] == 'Yes':
        return -1 # -1 means to match chatbot
    else:
        if len(Rest[n]) == 0: # if there's no person in the rest pool, we will return a chatbot
            return -1
        elif len(Like[n]) == 0 or len(Dislike[n]) == 0:
            return np.random.choice(Rest[n])
        else:
            X_train = []
            y_train = []
            for i in Like[n]:
                X_train.append(factor(n,i,db))
                y_train.append(1)
            for i in Dislike[n]:
                X_train.append(factor(n,i,db))
                y_train.append(0)
            model = LogisticRegression()
            model.fit(X_train, y_train)
            ans = Rest[n][0]
            temp = (model.intercept_+factor(n,ans,db)@model.coef_[0])[0]
            for j in Rest[n]:
                if (model.intercept_+factor(n,j,db)@model.coef_[0])[0]>temp:
                    temp = (model.intercept_+factor(n,j,db)@model.coef_[0])[0]
                    ans = j
            if np.random.choice([0, 1], p=[0.25, 0.75]) ==1:
                return ans
            else:
                return np.random.choice(Rest[n])
```
